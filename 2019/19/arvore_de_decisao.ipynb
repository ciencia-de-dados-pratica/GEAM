````python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
````

````python
caracteristicas = [[140,1],[130,1],[150,0],[170,0],[180,2],[175,2]]
````

````python
# rotulos = [0,0,1,1,2,2]
````

````python
rotulos = ['maca', 'maca', 'laranja', 'laranja', 'pera', 'pera']
````

````python
X = pd.DataFrame(caracteristicas,columns=['PESO','TEXTURA'])
````

````python
y = pd.DataFrame(rotulos, columns=['ROTULO'])
````

````python
final = pd.concat([X,y], axis=1)
final.sort_values(by='PESO')
````

````python
clf = DecisionTreeClassifier(random_state=42)
clf = clf.fit(caracteristicas, rotulos)
````

````python
clf.predict([[175,2]])
````

````python
plt.figure(figsize=[10,10])
plot_tree(clf, filled=True, feature_names = ['peso', 'textura'], class_names=['maa','laranja','pera']);
````

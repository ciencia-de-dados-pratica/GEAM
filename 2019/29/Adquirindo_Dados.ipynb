Lendo os dados da [API do TCE](https://api.tce.ce.gov.br/index.php/)

```python
import pandas as pd
```

```python
api = 'https://api.tce.ce.gov.br/index.php/sim/1_0/'
```

## Municípios e seus códigos

```python
municipios = pd.read_csv(api + 'municipios.csv', sep=';')
```

```python
municipios.head()
```

```python
municipios[municipios['nome_municipio'] == 'QUIXADA']
```

## Orgãos de Quixadá no ano de 2012

```python
orgaos_quixada = pd.read_csv(api + 'unidades_orcamentarias.csv?codigo_municipio=144&exercicio_orcamento=201200', sep=';')
```

```python
orgaos_quixada
```

## Agentes de Quixadá no ano de 2012

```python
agentes_quixada = pd.read_csv(api + 'agentes_publicos.csv?codigo_municipio=144&exercicio_orcamento=201200', sep=';')
```

```python
agentes_quixada.head()
```

### Ajustando o formato da tabela de agentes para ficar igual ao de orgãos

```python
agentes_quixada['codigo_orgao'] = agentes_quixada['codigo_orgao'].apply(lambda x: '{:02d}'.format(x))
```

## Descobrindo a quantidade de agentes que trabalham em cada orgão

```python
pd.merge(agentes_quixada, orgaos_quixada, on='codigo_orgao')['nome_unidade'].value_counts()
```

```python
from keras.applications import ResNet50
from keras.applications import InceptionV3
from keras.applications import Xception # TensorFlow ONLY
from keras.applications import VGG16
from keras.applications import VGG19
from keras.applications import imagenet_utils
from keras.applications.inception_v3 import preprocess_input
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
import numpy as np
import argparse
import cv2
import matplotlib.pyplot as plt
```

```python
def predizer(img, modelo):
    if modelo in ("3", "4"):
        inputShape = (299, 299)
        preprocess = preprocess_input
    else:
        inputShape = (224, 224)
        preprocess = imagenet_utils.preprocess_input

    img = cv2.resize(img, inputShape)
    # print(inputShape)


    Network = MODELS[modelo]
    model = Network(weights="imagenet")


    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess(img)

    preds = model.predict(img)
    P = imagenet_utils.decode_predictions(preds)

    print('\n\n\n')
    for (i, (imagenetID, label, prob)) in enumerate(P[0]):
    print("{}. {}: {:.2f}%".format(i + 1, label, prob * 100))
```

```python
MODELS = {
    "1": VGG16,
    "2": VGG19,
    "3": InceptionV3,
    "4": Xception, # TensorFlow ONLY
    "5": ResNet50
}
```

```python
imagem = input("Digite o nome da imagem (ex: imagem.jpg): \n")
# modelo = input("Digite o número do modelo que quer executar: \n1- VGG16\n2- VGG19\n3- InceptionV3\n4- Xception\n5- ResNet50\n")
# if(modelo != '1' and modelo != '2' and modelo != '3' and modelo != '4' and modelo != '5'):
#   print("modelo não existe")
# else:
#   try:
img = plt.imread(imagem)  
print("Modelo VGG16")
predizer(img, '1')
print("Modelo VGG19")
predizer(img, '2')
print("Modelo Inception")
predizer(img, '3')
print("Modelo Xception")
predizer(img, '4')
print("Modelo ResNet")
predizer(img, '5')
print("\n\n IMAGEM")
plt.imshow(img);
#   except:
#     print("Imagem não encontrada")
```
